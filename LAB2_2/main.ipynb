### Dataset Reading and Visualization

```python
import matplotlib.pyplot as plt
import torch
from utils import read_data, distortion, plot_images, plot_energy_overlap
from hopefield_net import HopfieldNet
```

```python
# read and stack data (second line is for transposing imgs)
X = read_data("data/")
X = torch.stack([img.reshape(32, 32).T.reshape(1024) for img in X])
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

fig.suptitle('Input Patterns', fontsize=16)

axs[0].imshow(X[0].reshape(32,32), cmap='gray')
axs[1].imshow(X[1].reshape(32,32), cmap='gray')
axs[2].imshow(X[2].reshape(32,32), cmap='gray')

plt.show()
```

### Network initialization

```python
net = HopfieldNet(X)
```

### Digit 0 - distortion and reconstructions

```python
d_0_1 = distortion(X[0], 0.05)
d_0_2 = distortion(X[0], 0.1)
d_0_3 = distortion(X[0], 0.25)
```

```python
r_0_1, r01_en, r01_ov = net.recall(d_0_1.squeeze(), X[0], 0.45)
r_0_2, r02_en, r02_ov = net.recall(d_0_2.squeeze(), X[0], 0.45)
r_0_3, r03_en, r03_ov = net.recall(d_0_3.squeeze(), X[0], 0.45)
```

```python
mse = torch.nn.functional.mse_loss(X[0], r_0_1).item()
print(f"MSE with 0.05 of noise: {mse:.4f}")
plot_images(X[0], d_0_1, r_0_1, 0.05, "digit_0")

mse = torch.nn.functional.mse_loss(X[0], r_0_2).item()
print(f"MSE with 0.1 of noise: {mse:.4f}")
plot_images(X[0], d_0_2, r_0_2, 0.1, "digit_0")

mse = torch.nn.functional.mse_loss(X[0].squeeze(), r_0_3).item()
print(f"MSE with 0.25 of noise: {mse:.4f}")
plot_images(X[0], d_0_3, r_0_3, 0.25, "digit_0")
```

```python
plot_energy_overlap(r01_en, r01_ov, 0.05, "digit_0")
plot_energy_overlap(r02_en, r02_ov, 0.1, "digit_0")
plot_energy_overlap(r03_en, r03_ov, 0.25, "digit_0")
```
